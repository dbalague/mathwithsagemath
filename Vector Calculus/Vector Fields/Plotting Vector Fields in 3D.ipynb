# Example 3

In this example we will plot the vector field $\displaystyle \mathbf{F}(x,y,z) = -\frac{x}{r}\,\mathbf{i} -\frac{y}{r}\,\mathbf{j} -\frac{z}{r}\,\mathbf{k}$ where $r = \sqrt{x^2+y^2+z^2}$.

If we try to use the function `plot_vector_field3d`, it will return us an error:

```python
y, z = var('y,z')
r(x,y,z) = sqrt(x^2+y^2+z^2)
F = [ -x/r, -y/r, -z/r ]

plot_vector_field3d(F,(x,-3,3),(y,-3,3),(z,-3,3),aspect_ratio=1)
```

In this case, since the vector field is not defined at the point $(0,0,0)$, we will plot it using two plots instead of one

```python
G1 = plot_vector_field3d(F,(x,-3,-0.01),(y,-3,3),(z,-3,3),aspect_ratio=1)
G2 = plot_vector_field3d(F,(x,0.01,3),(y,-3,3),(z,-3,3),aspect_ratio=1)

show(G1+G2)
```
